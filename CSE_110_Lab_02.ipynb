```python
#task23
degree=float(input('degrees Fahrenheit'))
celcius=((degree-32)*0.56)
if celcius<20:
  comment="Winter"
elif 20<=celcius<=25:
  comment="Autumn"
elif 25<celcius<30:
  comment="Spring"
else:
  comment="Summer"
print(f"{celcius:.2f } degrees C \n {comment}")
```
