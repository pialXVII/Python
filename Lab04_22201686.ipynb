```python
class Patient :
  def __init__(self,id,nam,boyos,rokto,next=None,prev=None):
    self.id =id
    self.name =nam
    self.age= boyos
    self.blood=rokto
    self.next =None
    self.prev= None

class WRM:
  def __init__(self):
    self.dh=Patient(None,None,None,None)
    self.size = 0
    self.next =self.dh
    self.prev = self.dh
    self.tail =self.dh

  def RegisterPatient(self,id, nam, boyos, rokto):
    Node = Patient(id, nam, boyos, rokto)
    self.tail.next = Node
    Node.prev = self.tail
    Node.next = self.dh
    self.dh.prev = Node
    self.tail = self.tail.next
    self.size +=1


  def ServePatient(self):
    if self.size !=0:
      serve = self.dh.next
      self.dh.next =serve.next
      serve.next.prev =serve.prev
      serve.next =None
      serve.prev = None
      self.size -=1
      print(f"Patient {serve.name} is being served.")

    else:
      print("No patients to serve.")

  def CancelAll(self):
    if self.size != 0:
      self.size  = 0
      self.dh.prev.next = None
      self.dh.next. prev = None
      self.dh.next = self.dh
      self.dh.prev = self.dh


  def CanDoctorGoHome(self):
    if self.size != 0:
      return False
    else:
      return True

  def ShowAllPatients(self):
    if self.size != 0:
      tail= self.dh.next
      while tail != self.dh:
        print(tail.name,end=" ")
        tail =tail.next
    else:
      print("No Patient Available!")

  def ReverseTheLine(self):
    if self.size !=0:

      newDH = Patient(None,None,None,None)
      newDH.next = newDH
      newDH.prev = newDH
      ntail= newDH

      while self.tail != self.dh:
        rev= self.tail.prev
        ntail.next =self.tail

        self.tail.prev = ntail
        self.tail.next = newDH
        newDH.prev =self.tail
        ntail =ntail.next
        self.tail= rev

      self.dh.prev= self.dh
      self.dh = newDH
      self.tail= ntail
      print("Line reversed successfully.")

    else:
      print("There is no patient to reverse!")








wrm = WRM()

while True:
    print("\nOptions:")
    print("1. Add Patient")
    print("2. Serve Patient")
    print("3. Show All Patients")
    print("4. Can Doctor Go Home?")
    print("5. Cancel All Appointments")
    print("6. Reverse The Line")
    print("7. Quit")
    print("===========================================")

    n = input("Choose option :")
    if n == '1':
        id = input("Enter patient ID: ")
        name = input("Enter patient name: ")
        age = input("Enter patient age: ")
        bloodgroup = input("Enter patient blood group: ")
        wrm.RegisterPatient(id, name, age, bloodgroup)
        print("Patient added successfully!")

    elif n == '2':
         wrm.ServePatient()


    elif n == '3':
         wrm.ShowAllPatients()


    elif n == '4':
        if wrm.CanDoctorGoHome():
            print("Yes, the doctor can go home.")
        else:
            print("No, there are patients waiting.")

    elif n == '5':
        wrm.CancelAll()
        print("All appointments canceled.")

    elif n == '6':
        wrm.ReverseTheLine()


    elif n == '7':
        print("Thank you")
        break

    else:
        print("Invalid choice. Please select a valid option.")
```
