```python
class Node:
  def __init__(self, elem = None, next = None):
    self.elem, self.next = elem, next

class Queue:
  def __init__(self):
    self.__start = None

  def enqueue(self, elem):
    nn = Node(elem)
    if self.__start == None:
      self.__start = nn
    else:
      temp = self.__start
      while temp.next != None:
        temp = temp.next
      temp.next = nn

  def dequeue(self):
    if self.__start == None:
      return None
    t = self.__start.elem
    self.__start = self.__start.next
    return t

  def peek(self):
    if self.__start != None:
      return self.__start.elem
```
