```python
class Node:
  def __init__(self, e, n, p):
    self.elem =e
    self.next = n
    self.prev = p
```

```python
class DoublyList:

  # Creates a Dummy Head for Dummy Headed Circular Doubly Circular Linked List.
  def __init__(self):
    self.dh=Node(None,None,None)
    self.dh.next=self.dh
    self.dh.prev=self.dh
    self.tail=self.dh




  #create a DHDCL from an array
  def create_list(self, l):
    self.tail=self.dh
    for i in range(len(l)):
      node=Node(l[i],None,None)
      self.tail.next=node
      self.tail.next.prev=self.tail
      self.tail=self.tail.next
      self.tail.next=self.dh
      self.dh.prev=self.tail




  # Counts the number of Nodes in the list and return the number
  def countNode(self):
    tail=self.dh
    tail=tail.next
    count=0
    while tail != self.dh:
      count+=1
      tail=tail.next
    return count


  # prints the elements in the list forward
  def forwardprint(self):
    tail=self.dh
    tail=tail.next
    while tail != self.dh:
      print(tail.elem,end=" ")
      tail=tail.next


  # prints the elements in the list backward
  def backwardprint(self):
    tail=self.dh
    tail=tail.prev
    while tail != self.dh:
      print(tail.elem,end=" ")
      tail=tail.prev


  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    tail=self.dh
    tail=tail.next
    count=0
    while tail != self.dh:
      if count ==idx:

        return tail
      count+=1
      tail=tail.next
    return None


  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    tail=self.dh
    tail=tail.next
    count=0
    while tail != self.dh:
      if tail.elem ==elem:

        return count
      count+=1
      tail=tail.next
    return -1


  # insert at specific index of the DHDCL
  def insert(self, elem, idx):
    node=Node(elem,None,None)
    #h= DoublyList()
    if 0< idx < self.countNode() :
      pre=self.nodeAt(idx-1)
      nex=self.nodeAt(idx)
      pre.next=node
      node.prev=pre
      node.next=nex
      nex.prev=node
    elif idx==0:
      pre=self.dh
      nex=self.nodeAt(idx)
      pre.next=node
      node.prev=pre
      node.next=nex
      nex.prev=node
    elif idx== self.countNode():
      pre=self.nodeAt(idx-1)
      nex= self.dh
      pre.next=node
      node.prev=pre
      node.next=nex
      nex.prev=node







  #removes the node of the specific index and returns the element of the node
  def remove(self, idx):
    rmv=self.nodeAt(idx)
    pre=self.nodeAt(idx-1)
    nxt=self.nodeAt(idx+1)
    if 0< idx < self.countNode()-1 :
      pre.next=nxt
      nxt.prev=pre
      rmv.next=None
      rmv.prev=None

    elif idx==0:
      self.dh.next=self.dh.next.next
      self.dh.next.prev.next=None
      self.dh.next.prev.prev=None
      self.dh.next.prev=self.dh
    else:
      pre.next=self.dh
      self.dh.prev= pre
      rmv.next=None
      rmv.prev=None

    return rmv.elem

```
