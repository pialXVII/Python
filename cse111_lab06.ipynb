```python
#task1
class Student:
  id=0
  def __init__(self,name,dept,age,cg):
    self.name=name
    self.dept=dept
    self.age=age
    self.cgpa=cg
    Student.id+=1
    self.id=Student.id

  def showDetails(self):
    print(f"""ID: {self.id}
Name: {self.name}
Department: {self.dept}
Age: {self.age}
CGPA: {self.cgpa}""")

  @classmethod
  def from_String(cls,s):
    x=s.split('-')
    obj=cls(x[0],x[1],x[2],x[3])
    return obj




s1 = Student("Samin", "CSE", 21, 3.91)
s1.showDetails()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.showDetails()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.showDetails()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.showDetails()
```

```python
#task2
class Assassin:
  total=0
  def __init__(self,name,rate):
    self.name=name
    self.rate=rate
    Assassin.total+=1

  def printDetails(self):
    print(f'''Name: {self.name}
Success rate: {self.rate}%
Total number of Assassin: {Assassin.total}''')

  @classmethod
  def failureRate(cls,name,fr):
    obj=cls(name,100-fr)
    return obj

  @classmethod
  def failurePercentage(cls,name,fp):
    fp=int(fp[:len(fp)-1])
    obj=cls(name,100-fp)
    return obj



john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()

```

```python
#task6
class Laptop:
  laptopCount=0
  def __init__(self,name,count):
    self.name=name
    self.count=count
    Laptop.laptopCount+=count

  @staticmethod
  def advantage():
    print('Laptops are portable')

  @classmethod
  def resetCount(cls):
    cls.laptopCount=0



lenovo = Laptop("Lenovo", 5)
dell = Laptop("Dell", 7)
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
#task7
class Cat:
  Number_of_cats = 0

  def __init__(self, colour, action):
    self.colour = colour
    self.action = action
    Cat.Number_of_cats += 1

  @classmethod
  def no_parameter(cls):
    return cls("White", "sitting")

  @classmethod
  def first_parameter(cls, colour):
    return cls(colour, "sitting")

  @classmethod
  def second_parameter(cls, action):
    return cls("Grey", action)

  def changeColor(self, colour):
    self.colour = colour

  def printCat(self):
    print(f"{self.colour} cat is {self.action}")

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#task8
import math
class Cylinder:
  height=18
  radius=5
  def __init__(self,r,h):
    self.height=h
    self.radius=r
    print(f"""Default radius={Cylinder.radius} and height={Cylinder.height}.
Updated: radius={self.radius} and height={self.height}.""")
    Cylinder.radius=self.radius
    Cylinder.height=self.height

  @staticmethod
  def area(r,h):
    a= (2*math.pi*r*r) + (2*math.pi*r*h)
    print(f"Area: {a}")

  @staticmethod
  def volume(r,h):
    v=math.pi*r*r*h
    print(f"Volume:  {v}")

  @classmethod
  def swap(cls,r,h):
    obj=cls(h,r)
    return obj

  @classmethod
  def changeFormat(cls,s):
    x=s.split('-')
    obj=cls(float(x[0]),float(x[1]))
    return obj






c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)

```

```python

#task10
class SultansDine:
  totalBranch=0
  totalSell=0
  total=[]

  def __init__(self,name):
    self.name=name
    self.sell=0
    SultansDine.totalBranch+=1
    SultansDine.total.append(self.name)


  def sellQuantity(self,n):
    if n <10:

      SultansDine.totalSell+= 300*n
      self.sell= 300*n
      SultansDine.total.append(self.sell)
    elif n <20:

      SultansDine.totalSell+= 350*n
      self.sell= 350*n
      SultansDine.total.append(self.sell)
    else:


      SultansDine.totalSell+= 400*n
      self.sell = 400*n
      SultansDine.total.append(self.sell)


  def branchInformation(self):
    print(f"""Branch Name: {self.name}
Branch Sell: {self.sell} Taka""")

  @classmethod
  def details(cls):
    print(f"""Total Number of branch(s):{cls.totalBranch}
Total Sell: {cls.totalSell} Taka""")
    if len(cls.total)!=0:

      for i in range(0,len(cls.total)-1,2):
        print(f'Branch Name: {cls.total[i]}, Branch Sell: {cls.total[i+1]} Taka')
        percent=( cls.total[i+1] / cls.totalSell )*100
        print(f"Branch consists of total sell's: {percent:.2f}%")








SultansDine.details()
print('########################')
dhanmondi = SultansDine('Dhanmondi')
dhanmondi.sellQuantity(25)
dhanmondi.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
