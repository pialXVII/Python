```python
def sumDigits(n):
  if n ==0:
    return 0

  x=n%10

  z= x+ sumDigits(n//10)

  return z


print(sumDigits(126))
print(sumDigits(49))
print(sumDigits(12))
```

```python
def bunnyEars2(n):
  if n== 0:
    return 0
  if n%2==0:
    total = 3+ bunnyEars2(n-1)
  else:
    total = 2+ bunnyEars2(n-1)

  return total


print(bunnyEars2(0))
print(bunnyEars2(1))
print(bunnyEars2(2))
```

```python
def count7(n):
  if n== 0:
    return 0
  x= n%10
  if x == 7:
    total = 1+count7(n//10)
  else:
    total = 0+count7(n//10)

  return total






print(count7(717))
print(count7(7))
print(count7(123))
```

```python
def countX(s):
  if s== "":
    return 0
  x= s[0]
  if x == "x":
    total = 1+countX(s[1:])
  else:
    total = 0+countX(s[1:])

  return total






print(countX("xxhixx"))
print(countX("xhixhix"))
print(countX("hi"))

```

```python
def changePi(s):
  if s=="":
    return ""
  if len(s) > 1 and s[0] =="p" and s[1]=="i":
    x="3.14"+ changePi(s[2:])
  else:
    x=s[0]+changePi(s[1:])
  return x

print(changePi("xpix"))
print(changePi("pipi"))
print(changePi("pip"))
```

```python
def countHi2(s):
  if s=="":
    return 0
  if s[0]=="x":

    total = 0 + countHi2(s[3:])
  elif len(s)>1 and s[0] =="h" and s[1] == "i":
    total = 1 + countHi2(s[2:])
  else:
    total = 0 + countHi2(s[1:])
  return total

print(countHi2("ahixhi"))
print(countHi2("ahibhi"))
print(countHi2("xhixhi"))
```
